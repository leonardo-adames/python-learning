Los booleanos representan uno de dos valores: o .`True` o `False`


Valores booleanos
En programación, a menudo necesitas saber si una expresión es o .TrueFalse


Puedes evaluar cualquier expresión en Python y obtener una de dos respuestas, o .TrueFalse


Cuando comparas dos valores, la expresión se evalúa y Python devuelve la respuesta booleana.

```python
print("Ejemplo:")

print("10 > 9", 10 > 9)
print("10 == 9", 10 == 9)
print("10 < 9", 10 < 9)
```

____________________________________________________________________
Cuando ejecutas una condición en una sentencia if, 

Python devuelve o :TrueFalse

Ejemplo

Imprime un mensaje basado en si la condición es o :TrueFalse

```python
a = 200
b = 33

if b > a:
  print("b is greater than a")
else:
  print("b is not greater than a")
```
