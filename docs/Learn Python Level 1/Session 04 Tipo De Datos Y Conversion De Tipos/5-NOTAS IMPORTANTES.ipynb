# CUANO USAR Y NO USAR LOS TIPOS DE  ESTRUCTURAS

DE DATOS EN PYTHON (LISTAS, TUPLAS, DICCIONARIOS, CONJUNTOS, CONJ. INMUTABLES, RANGO, BYTRARRAY, MEMORY VIEW)

LISTAS: list[]
--------------
Es recomendable usar las listas cuando neccesites
una secuencia que vaya a
espesificar, agregar o quitar elementos.


TUPLAS: tuple()
---------------
Cuando necesites una secuencia de lisa fija
ejemplo(COORDENADAS, COLORES RGB), etc.

DICCIONARIO: dict{}
-------------------
cuando necesites ordnar valores mediante palabras claves

CONJUNTOS: set{}
-----------------
Para eliminar duplicados o hacer operaciones matemáticas
de conjunto.

CONJ.INMUTABLE: frozenset ([])
-------------------------------
Cuando necesitas un conjunto que NO pueda cambiarse.

Cuando requieres usar un conjunto como clave de diccionario (solo las cosas inmutables pueden ser claves).

Para representar colecciones que deben protegerse contra cambios accidentales.

Cuando usas conjuntos dentro de otros conjuntos (solo se permiten elementos inmutables).

ejemplo:

colores = frozenset(["rojo", "verde", "azul"])


RANGO: (range)
------------
Para ciclos for donde necesitas un contador.

Para generar secuencias de números eficientemente (no ocupa memoria grande).

Cuando necesitas iterar sobre una cantidad grande de números sin almacenar todos.

Para crear listas ordenadas solo si es necesario convertirlo.

Ejemplo:

for i in range(1, 6):

    print(i)


BYTEARRAY: (bytearray)
--------------------

Cuando necesitas datos binarios modificables.

Lectura y escritura de archivos binarios.

Comunicación con dispositivos, sensores, sockets.

Criptografía, compresión o manipulación de bytes.

Es como bytes, pero mutable.

Ejemplo:

data = bytearray([65, 66, 67])

data[0] = 97   # modifica a 'a'


MEMORYVIEW: (memoryview)
--------------------

Cuando quieres acceder o modificar datos binarios sin copiarlos.

Para trabajar con archivos, buffers, imágenes o streams de gran tamaño.

Para mejorar rendimiento cuando se trabaja con mucho contenido en bytes.

Para manipular partes de datos sin duplicar memoria.

Ejemplo:

data = bytearray(b"hola")

mv = memoryview(data)

mv[0] = 72  # cambia 'h' por 'H'
