Instrumentos de cuerda (STRINGS)
Las cadenas en Python están rodeadas de comillas simples o comillas dobles.


`'Hola'` es lo mismo que `'Hola'`.


Puedes mostrar una cadena literal con la función: `print()`

Ejemplo

```python
print("Hello")
print('Hello')
```

## Citas dentro de las citas

Puedes usar comillas dentro de una cadena, siempre que no coincidan con las comillas que rodean la cadena:

Ejemplo:

```python
print("It's alright")
print("He is called 'Johnny'")
print('He is called "Johnny"')
```

## Asignar cadena a una variable

Asignar una cadena a una variable se hace con el nombre de la variable seguido de un signo igual y la cadena:

```python
print("Ejemplo")


a = "Hello"
print(a)
```

## Cuerdas multilínea

Puedes asignar una cadena multilínea a una variable usando tres comillas:

```python
print("Ejemplo")
print("Puedes usar tres comillas dobles:")
print("-------------------------")

a = """Lorem ipsum dolor sit amet,
consectetur adipiscing elit,
sed do eiusmod tempor incididunt
ut labore et dolore magna aliqua."""
print("print (a) =", (a))
```

```python
print("O tres citas simples:")

print("Ejemplo")
print("-------------------------")
a = '''Lorem ipsum dolor sit amet,
consectetur adipiscing elit,
sed do eiusmod tempor incididunt
ut labore et dolore magna aliqua.'''
print("print(a) =",a)
```

Nota: en el resultado, los saltos de línea se insertan en la misma posición que en el código.

EL RESULTADO ES  EL MISMO CON CUALQUIERA DE LAS DOS.

## Indexación de cuerdas 

Como muchos otros lenguajes de programación populares, las cadenas en Python son matrices de caracteres unicode.


Sin embargo, Python no tiene un tipo de dato de carácter, un solo carácter es simplemente una cadena con una longitud de 1.


Los corchetes pueden usarse para acceder a elementos de la cuerda.

```python
enunciado = '''Ejemplo
Coloca al personaje en la posición 1 (recuerda que el primer personaje tiene el posición 0):'''
print(enunciado)

a = "Hello, World!"
print(a[1]) #el resultado es (e)
```

## Bucle a través de una cuerda

Como las cadenas son arrays, 
podemos recorrer los caracteres de una cadena, con un `bucle.'for'`


```python
print("Ejemplo:")
print("Haz un bucle entre las letras de la palabra 'banana':")

for x in "banana":
  print(x) # x= 'a'
```

## LONGITUD DE CUERDA (STRING)

Para obtener la longitud de una cadena, usa la función.len()


```python
text = '''Ejemplo
La función devuelve la longitud de una cadena:len()'''
print(text)

a = "Hello, World!"
print("Longitud de 'a' = ",len(a))  # Mide la longitud de la cadena y contabiliza sus caracteres
```

## Cuerda de control

Para comprobar si una frase o carácter concreto está presente en una cadena, podemos usar La palabra clave `.in`

Ejemplo

```python
print("Comprueba si 'gratis' aparece en el siguiente texto:")

txt = "The best things in life are free!"
print("free" in txt)
```

```python
#Úsalo en una afirmación:if

#Ejemplo
#Imprimir solo si está "libre":

txt = "The best things in life are free!"
if "free" in txt:
  print("Yes, 'free' is present.")
```

## Comprueba si no.

Para comprobar si una frase o carácter concreto NO está presente en una cadena, 

podemos usar La palabra clave `.not in`

Ejemplo



```python
#Comprueba si "caro" NO está presente en el siguiente texto:

txt = "The best things in life are free!"
print("expensive" not in txt)

print("------------------------------------------------------")
#Úsalo en una afirmación:if
#Ejemplo
#imprimir solo si "caro" NO está presente:

txt = "The best things in life are free!"
if "expensive" not in txt:
  print("No, 'expensive' is NOT present.")
```
